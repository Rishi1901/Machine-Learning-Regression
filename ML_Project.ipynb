## changing the college column into integer by diving colleges into groups 

```python
for item in df['College']:
    if item in tier1:
        df['College'].replace(item,1,inplace=True)
    elif item in tier2:
        df['College'].replace(item,2,inplace=True)
    elif item in tier3:
        df['College'].replace(item,3,inplace=True)
```

```python
df.head()
```

```python
metro = city['Metrio City'].tolist()
non_metro=city['non-metro cities'].tolist()
```

## Changing city into int by converting into two categories 

```python
for item in df['City']:
    if item in metro:
        df['City'].replace(item,'metro',inplace=True)
    elif item in non_metro:
        df['City'].replace(item,'non_metro',inplace=True)
```
